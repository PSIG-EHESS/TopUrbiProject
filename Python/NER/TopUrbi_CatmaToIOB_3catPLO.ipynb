````python
def transform_input_to_output(input_data):
    entities = []
    for item in input_data:
        #print(item)
        start_offset = item["start_offset"]
        end_offset = item["end_offset"]
        tag_name = item["tag_name"]
        text_cont = item["annotation"]
        if start_offset is not None and end_offset is not None and tag_name is not None:
            entities.append((start_offset, end_offset, tag_name, text_cont))

    output = {'entities': entities}
    return output
````

Creacion del formato para importar les anotaciones en LabelStudio y poder luego transformarlas al format CONLL2023 (IOB) requerido por Spacy, el formato JSON de entrada es este :https://labelstud.io/guide/predictions#Import-span-pre-annotations-for-text.

````python
count = 0
results = []
for ent in spacy_json['entities']:
  results.append({
            "id": "ann_"+str(count),
            "from_name": "label",
            "to_name": "text",
            "type": "labels",
            "value": {
              "start": ent[0],
              "end": ent[1],
              "score": 1,
              "text": ent[3],
              "labels": [
                ent[2]
              ]
            }
          })
  count = count + 1

ann = [
    {
    "data": {
      "text": spacy_json['text']
    },
    "predictions": [
      {
        "model_version": "one",
        "score": 1,
        "result" : results
      }
    ]
  }
]
````

````python
count = 0
results = []
for ent in spacy_eval_json['entities']:
  results.append({
            "id": "ann_"+str(count),
            "from_name": "label",
            "to_name": "text",
            "type": "labels",
            "value": {
              "start": ent[0],
              "end": ent[1],
              "score": 1,
              "text": ent[3],
              "labels": [
                ent[2]
              ]
            }
          })
  count = count + 1

ann_eval = [
    {
    "data": {
      "text": spacy_eval_json['text']
    },
    "predictions": [
      {
        "model_version": "one",
        "score": 1,
        "result" : results
      }
    ]
  }
]
````

Para cada archivo IOB, remplazamiento manuel con vérification en notepad++ de lo siguiente, de manera segmentar el texto (par crear los Doc de Spacy), necesario para el entrenamiento :

```
1) "-X- _" hacia "|"

2) "@|O
@|O
@|O
A" hacia "A"

3) @|O
@|O
@|O
#|O hacia ""

4) "§§§|O\n" hacia "§§§|O\n-DOCSTART- -X- O O\n"

5) "|" hacia " " (es un espacio)
```
