Crear json a partir del data frame

````python
json_data_A = df_A.to_json(orient = 'records')
json_data_B = df_B.to_json(orient = 'records')
json_data_C = df_C.to_json(orient = 'records')
````

````python
json_data_obj_A = list(eval(json_data_A))
json_data_obj_B = list(eval(json_data_B))
json_data_obj_C = list(eval(json_data_C))
````

````python
json_data_obj_B
````

````python
def transform_input_to_output(input_data):
    entities = []
    for item in input_data:
        #print(item)
        start_offset = item["start_offset"]
        end_offset = item["end_offset"]
        tag_name = item["tag_name"]
        text_cont = item["annotation"]
        if start_offset is not None and end_offset is not None and tag_name is not None:
            entities.append((start_offset, end_offset, tag_name, text_cont))

    output = {'entities': entities}
    return output
````

````python
spacy_json_A = transform_input_to_output(json_data_obj_A)
spacy_json_B = transform_input_to_output(json_data_obj_B)
spacy_json_C = transform_input_to_output(json_data_obj_C)
````

Para cada archivo IOB, remplazamiento manuel con vérification en notepad++ de lo siguiente, de manera segmentar el texto (par crear los Doc de Spacy), necesario para el entrenamiento :

```
1) "-X- _" hacia "|"

2) "@|O
@|O
@|O
A" hacia "A"

3) @|O
@|O
@|O
#|O hacia ""

4) "§§§|O\n" hacia "§§§|O\n-DOCSTART- -X- O O\n"

5) "|" hacia " " (es un espacio)

6) los puntos al final de un nombre se separa en otra linea
```
