```python
import os
import spacy
import spacy_transformers
from spacy.tokens import Doc

model_3catLoc = gdrive_path + 'models/model_spacy_3catLoc_gpu_trf_es/model-best'

nlp = spacy.load(model_3catLoc)

# Dossier contenant les fichiers texte
dossier_textes = gdrive_path + 'dico_alcedo_txt/'

# Fonction pour baliser le texte avec des tags XML
def baliser_entites_xml(texte, entites):
  deja_vu = []
  for entite in entites:
    if (entite.text not in deja_vu): #en cas de doublons, on donne la préférence à la première occurrence
      #print(entite.label_ + " " + entite.text + " " + str(entite.start) + " " + str(entite.end))
      tag = f'<{entite.label_}>{entite.text}</{entite.label_}>'
      deja_vu.append(entite.text)
      texte = texte.replace(entite.text, tag)
  return texte

# Lire le contenu du fichier
for fichier in os.listdir(dossier_textes):
  if fichier.endswith('.source.txt'):
    chemin_fichier = os.path.join(dossier_textes, fichier)
    with open(chemin_fichier, 'r', encoding='utf-8') as file:
      contenu = file.read()
      paragraphes = contenu.split('\n\n')  # Changer le séparateur si nécessaire
      texte_balise = ''
      for paragraphe in paragraphes:
        doc = nlp(paragraphe)
        texte_balise_seul = baliser_entites_xml(paragraphe, doc.ents)
        #print(texte_balise_seul)
        texte_balise += texte_balise_seul + '\n\n'

      chemin_resultat = os.path.splitext(chemin_fichier)[0] + '_tag3catLoc.txt'
      with open(chemin_resultat, 'w', encoding='utf-8') as result_file:
        result_file.write(texte_balise)
      print(f'Le fichier {fichier} a été traité avec succès. Résultat sauvegardé dans {chemin_resultat}')
```

4) LOC, PER, ORG + 3 cat geo pero con controles

```python
import os
import spacy
import spacy_transformers
from spacy.tokens import Doc

model_3catPLO = gdrive_path + 'models/model_spacy_3catPLO_gpu_trf_es/model-best'

nlp = spacy.load(model_3catPLO)

# Dossier contenant les fichiers texte
dossier_textes = gdrive_path + 'dico_alcedo_txt/'

# Fonction pour baliser le texte avec des tags XML
def baliser_entites_xml(texte, entites):
  deja_vu = []
  deja_vu_offsets = []
  sous_strings_interdits = ["<", ">"]
  for entite in entites:
    if (entite.text not in deja_vu and not any(sous_string in entite.text for sous_string in sous_strings_interdits)):
      #en cas de doublons, on donne la préférence à la première occurrence ainsi qu'au tag déjà présent
      print(entite.label_ + " " + entite.text + " " + str(entite.start) + " " + str(entite.end))
      tag = f'<{entite.label_}>{entite.text}</{entite.label_}>'
      deja_vu.append(entite.text)
      deja_vu_offsets.append((entite.start, entite.end))
      texte = texte.replace(entite.text, tag)
    else:
      print(entite.label_ + " " + entite.text)
  return texte

# Lire le contenu du fichier
for fichier in os.listdir(dossier_textes):
  if fichier.endswith('_tag3catLoc.txt'):
    chemin_fichier = os.path.join(dossier_textes, fichier)
    with open(chemin_fichier, 'r', encoding='utf-8') as file:
      contenu = file.read()
      paragraphes = contenu.split('\n\n')  # Changer le séparateur si nécessaire
      texte_balise = ''
      for paragraphe in paragraphes:
        doc = nlp(paragraphe)
        texte_balise_seul = baliser_entites_xml(paragraphe, doc.ents)
        #print(texte_balise_seul)
        texte_balise += texte_balise_seul + '\n\n'

      chemin_resultat = os.path.splitext(chemin_fichier)[0] + '_tagPLO.txt'
      with open(chemin_resultat, 'w', encoding='utf-8') as result_file:
        result_file.write(texte_balise)
      print(f'Le fichier {fichier} a été traité avec succès. Résultat sauvegardé dans {chemin_resultat}')
```
